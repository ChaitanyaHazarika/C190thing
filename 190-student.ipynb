```python
#import the libraries
import pandas as pd
from matplotlib import pyplot as plt

bmi_df= pd.read_csv('bmi.csv')
bmi_df

#Task 1
#Read the bmi.csv
```

```python
#Task 2
#Data is sorted in descending order in accordance with BMI value
#Find the top 5 age group where the BMI value is the highest, and plot a bar graph out of it

t5=bmi_df.head(5)
name= t5['Age']
number= t5["BMI"]
plt.xlabel("Age")
plt.xticks(rotation= 'vertical')
plt.ylabel("BMI")

label= name
value= number
print(label)
print(value)
plt.bar(label, value, width= 0.5, color=("Blue", "Green", "Black", "Pink", "Red"))
```
